## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
r=requests.get('https://google.com')
rr=r.text
s=BeautifulSoup(rr)
```

```python
print(s)
```

```python
s.find('title')
```

```python
print(type(rr))
print(type(s))
```

## Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq
g = Grab()
resp = g.go('https://google.com')
resp.body
```

```python
doc = pq(resp.body)
title = doc('title')
print(type(title), title.text())
```

```python
print(type(g))
print(type(resp))
print(type(resp.body))
print(type(doc))
print(type(title))
```
