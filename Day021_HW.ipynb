## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
print(time.localtime())
print(time.localtime().tm_year)
print(type(time.localtime().tm_year))
```

```python
time_range=[str(time.localtime().tm_hour-i) for i in range(3)]
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text[-5]+d.find(class_="date").text[-4] in time_range:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        continue
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
year=str(time.localtime().tm_year)
month=str(time.localtime().tm_mon)
day=str(time.localtime().tm_mday-3)
new_url="https://www.ettoday.net/news/news-list"+'-'+year+'-'+month+'-'+day+'-0'+".htm"

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(new_url)
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
# 每1秒鐘自動往下滑

import time
for i in range(10):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
```
