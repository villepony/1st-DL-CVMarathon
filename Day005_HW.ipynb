```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
rr=r.text
json.loads(rr)
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
j=json.loads(rr)
print(len(j))
print(j[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for i in range(len(j)):
    print('æ¨™é¡Œï¼š'+j[i]["title"])
    print('è²¼æ–‡æ™‚é–“ï¼š'+j[i]["createdAt"])
    print('ç•™è¨€äººæ•¸ï¼š'+str(j[i]["commentCount"]))
    print('æŒ‰è®šäººæ•¸ï¼š'+str(j[i]["likeCount"]))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
rp=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
rrp=rp.text
jp=json.loads(rrp)
popular_comments_count=0
popular_likes_count=0
for i in range(len(jp)):
    popular_comments_count+=jp[i]["commentCount"]
    popular_likes_count+=jp[i]["likeCount"]
print(popular_comments_count/len(jp))
print(popular_likes_count/len(jp))
```

```python
rnp=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
rrnp=rnp.text
jnp=json.loads(rrnp)
nonpopular_comments_count=0
nonpopular_likes_count=0
for i in range(len(jnp)):
    nonpopular_comments_count+=jnp[i]["commentCount"]
    nonpopular_likes_count+=jnp[i]["likeCount"]
print(nonpopular_comments_count/len(jnp))
print(nonpopular_likes_count/len(jnp))
```
