### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
html_source

# 取得資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
course = []
    
for i in soup.find_all('h4', class_='title marg-t-20 marg-b-10'):
    course.append(i.text)

print(course)
```

```python
browser.close()
```

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
html_source
```

```python
# 1. 先取出全部有幾頁
pages=soup.find('div', class_="pagination-container")
pages.find_all('a')
```

```python
for i in pages.find_all('a'):
    print(i.text)
```

```python
pages.find_all('a')[-2].text
```

```python
# 2. 每頁都取出資料放入 courses

courses = []
for i in range(1,int(pages.find_all('a')[-2].text)+1):
    browser = webdriver.Chrome(executable_path='chromedriver')
    browser.get("https://hahow.in/courses?page="+str(i))
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    for i in soup.find_all('h4', class_='title marg-t-20 marg-b-10'):
        courses.append(i.text)
    browser.close()
```

```python
# 3. done!

print(courses)
```
