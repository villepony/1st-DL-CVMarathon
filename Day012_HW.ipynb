### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
type=set()
for d in soup.find(class_="part_list_2").find_all('h3'):
    type.add(d.find_all('em')[-1].text)
type=list(type.copy())
print(type)
print(type[0])
```

```python
for i in range(len(type)):
    print("a"+str(i)+"=[]")
for i in range(len(type)):
    print("a"+str(i)+",")
```

```python
a0=[]
a1=[]
a2=[]
a3=[]
a4=[]
a5=[]
a6=[]
a7=[]
a8=[]
a9=[]
a10=[]
a11=[]
a12=[]
a13=[]
a14=[]
a15=[]
a16=[]
a17=[]
typestoapd=[a0,
a1,
a2,
a3,
a4,
a5,
a6,
a7,
a8,
a9,
a10,
a11,
a12,
a13,
a14,
a15,
a16,
a17]
for d in soup.find(class_="part_list_2").find_all('h3'):
    for i in type:
        if d.find_all('em')[-1].text != i:
            continue
        else:
            typestoapd[type.index(i)].append(d.find(class_="date").text+' '+d.find_all('a')[0].text)
print(a0)
```

```python
ii={}
for i in type:
    ii[i]=typestoapd[type.index(i)]
print(ii)
```

### ③ 哪一個類別的文章最多

```python
for i in ii:
    print(i+': '+str(len(ii[i])))
```
